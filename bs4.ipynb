# beautifulSoup类的基本元素

|基本元素|说明|输出类型|
|--|--|--|
|Tag|标签，最基本的信息组织元素，分别用<></>表明开头与结尾|string|
|Name|标签的名称，如p标签，标签的名字为p,格式：<tag>.name|string|
|Atttibutes|标签的属性，字典形式组织，格式：<tag>.attrs|字典|
|navigableString|标签的非属性的字符串，<>..</>中的字符串,格式：<tag>,string|String|
|Comment|标签当中的字符串的注释部分，一种特殊的commit类型|String|

```python
#创建一个bs类
import requests
from bs4 import BeautifulSoup
r = requests.get('http://www.chamiji.com/countryrank').text
soup = BeautifulSoup(r,'html.parser')

#prettify()方法，更好的显示标签，会为标签添加换行，更好的阅读。
content = soup.tbody.prettify()
print(content)
    
```

## 标签的下行遍历

|属性|说明|
|--|--|
|.contents|子节点的列表，将tag表亲的所有子节点存入列表当中|
|.children|子节点的迭代类型，与.content类似，用于循环遍历子节点|
|.descendants|子节点的迭代类型,包含所有的子孙节点，用于循环遍历|

## 标签的上行遍历
|属性|说明|
|--|--|
|.parent|节点的父节点|
|.parent|所有的父节点迭代类型，用于循环的遍历父节点|

## 标签的平行遍历

平行标签的是指在标签树当中具有同一个父节点且在同一个平行位置的标签。

|属性|说明|
|---|---|
|.next_sibling|返回按照html文本顺序的下一个平行节点标签|
|.prevrious_sibling|返回按照html文本顺序的上一个平行节点标签|
|.next_siblings|迭代类型，返回按照html文本顺序的后续所有平行节点标签|
|.prevrious_siblings|迭代类型，返回按照html文本顺序的前续所有平行节点标签|

### 示例：

```python
import requests
from bs4 import BeautifulSoup

def get_content(url):
    try:    
        r = requests.get(url)
        r.raise_for_status()
        return r.text
    except:
        return ''
def soup(html):
    soup = BeautifulSoup(html,'html.parser')
    return soup.a.prettify()
def main():
    url = 'http://www.chamiji.com/countryrank'
    html = get_content(url)
    print(soup(html))
main()
```

## 方法函数

|方法函数|解释|返回值类型|参数|
|--|--|--|--|
|<>.findall(name,attrs,recursive,string,\*\*)kwargs)|查找html当中符合要求的值|列表类型|name:标签的名字<br>attrs：标签的属性<br> rescursive:是否对子孙节点进行查找，默认是true<br> string:标签内的字符串进行匹配<b>|
|<>.find_parent()|在父节点搜索且返回一个结果，|字符串类型|
|<>.find_parents()|在父节点搜索且返回一个结果，|列表类型|
|<>.find_next_sibling()|在后序平行节点搜索且返回一个结果，|字符串类型|
|<>.find_next_siblings()|在后序平行点搜索且返回一个结果，|列表类型|
|<>.find_previous_sibling()|在前序平行节点搜索且返回一个结果，|字符串类型|
|<>.find_previous_sibling()|前在前序平行节点搜索且返回一个结果，|列表类型|
|--|---|--|其他方法与finall（）的参数信息相同|



```python
import requests
import bs4
from bs4 import BeautifulSoup4

def getHTMLText(url):
    try:
        r = requests.get(url)
        r.raise_for_status()
        return r.text
    except:
        return ''
    
def fillUnivList(ulist,html):
    soup = BeautifulSoup(html,'html.parser')
    for tr in soup.find('tbody').children:
        tds = tr('td')
        ulist.append(tds[0].string,tds[1].string,tds[2].string)
        
def printUnivList(ulist,num):
    tplt = "{0:^10}\t{1:{3}^10}\t{2:^10}"
    print(tplt.format("排名","学校","总分",chr(12288)))
    for i in range(num)：
        u = ulist[i]
        print(tplt.format(u[0],u[1],u[2],chr(12288)))

def main():
    url = 'http://www.zuihaodaxue.com/zuihaodaxuepaiming2016.html'
    uinfo = []
    html = getHTMLText(url)
    fillUnivList(uinfo,html)
    printUnivList(uinfo,20)
if __name__ == '__main__':
    main()
```
